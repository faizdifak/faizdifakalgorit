```python
tahun = int(input("Masukkan tahun: "))
if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
    print(f"{tahun} adalah tahun kabisat.")
else:
    print(f"{tahun} adalah bukan tahun kabisat")

```

```python
angka1 = float(input("Masukkan angka pertama: "))
angka2 = float(input("Masukkan angka kedua: "))
angka3 = float(input("Masukkan angka ketiga: "))

if angka1 >= angka2 and angka1 >= angka3:
    terbesar = angka1
elif angka2 >= angka1 and angka2 >= angka3:
    terbesar = angka2
else:
    terbesar = angka3

print(f"Angka pertama: {angka1}")
print(f"Angka kedua: {angka2}")
print(f"Angka ketiga: {angka3}")
print(f"Angka terbesar di antara ketiga angka yang dimasukkan adalah: {terbesar}")
```

```python
# @title Default title text
jenis_konsumen = input("Apakah Anda member atau non member? (member/non member): ").strip().lower()
total_belanja = float(input("Masukkan total belanja: "))

if total_belanja > 200000:
    diskon = 0.04
elif total_belanja > 100000:
    diskon =
    0.03
else:
    diskon = 0.0

if jenis_konsumen == "member":
    diskon += 0.02

jumlah_diskon = total_belanja * diskon
jumlah_bayar = total_belanja -jumlah_diskon

print(f"Total belanja: Rp{total_belanja:.2f}")
print(f"Jumlah diskon yang diperoleh: Rp{jumlah_diskon:.2f}")
print(f"Jumlah yang harus dibayarkan: Rp{jumlah_bayar:.2f}")
```

```python
username = input("Masukan username: ")
email = input("Masukan email: ")
password = input("Masukan password: ")

if not username:
  print("Username tidak boleh kosong.")
elif "@" not in email:
  print("Email harus mengandung '@'.")
elif len(password) <= 8:
  print("Password harus memiliki lebih dari 8 karakter.")
else:
  print("User baru berhasil dibuat!")
  print(f"Username: {username}")
  print(f'Email: {email}')
  print(f"Password: {password}")
```
