```python
num = int(input("Masukan Angka "))
for i in range(1, num + 1):
  print(" " * (num - i), end=" ")
  for j in range(1, i + 1):
    print("*", end=" ")
  print()
```

```python
triangle = int(input("Masukkan jumlah baris: "))
rows = []
for i in range(triangle):
    row =[1] * (i + 1)
    rows.append(row)

    for j in range(1, i):
      row[j] = rows[i - 1][j - 1] + rows[i - 1][j]

for i in range(triangle):
  print(' ' * (triangle - i - 1) ,end= ' ')

  for j in range(i + 1):
    print(rows[i][j], end=' ')
  print()
```

```python
def generate_pascal_triangle(n):
  triangle = [[1 for _ in range(i+1)] for i in range(n)]
  for i in range(2, n):
    for j in range(1, i):
      triangle[i][j] = triangle[i-1][j-1] + triangle[i-1][j]
    return triangle

def sum_of_row(triangle, row):
  return sum(triangle[row-1])

n = int(input("Masukkan angka: "))
triangle = generate_pascal_triangle(n)
print("Baris segitiga pascal ke:")
print(" ".join(str(num) for num in triangle[n-1]))
print("Jumlah angka pada baris segitiga pascal ke:", sum_of_row(triangle, n))
```
