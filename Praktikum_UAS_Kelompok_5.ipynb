```python
# kelas buku dengan atribut judul, pengarang, tahun terbit, jumlah stok
class Buku:
    # Inisialisasi atribut buku
    def __init__(self, judul, pengarang, tahun_terbit, jumlah_stok):
      self.judul = judul
      self.pengarang = pengarang
      self.tahun_terbit = tahun_terbit
      self.jumlah_stok = jumlah_stok

    # metode untuk pinjam buku
    def pinjam(self):
         # Untuk mengecek Apakah stok buku masih ada
         if self.jumlah_stok > 0:
            self.jumlah_stok -= 1
            print(f"Buku {self.judul} berhasil dipinjam.")
         else:
            # Jika stok habis, maka akan menampilkan pesan error stok buku habis
            raise ValueError(f"Stok buku habis.")

    # metode untuk kembalikan buku
    def kembalikan(self):
        self.jumlah_stok += 1
        print(f"Buku {self.judul} berhasil dikembalikan.")

# list untuk menyimpan daftar buku
daftar_buku = []
# Dictionary untuk menyimpan data buku berdasarkan judul
data_buku = {}

# fungsi untuk menampilkan semua buku
def tampilkan_semua_buku():
    # Jika daftar buku kosong, maka cetak pesan tidak ada buku yang tersedia
    if not daftar_buku:
        print("Tidak ada buku yang tersedia.")
        return
    # Cetak Daftar buku beserta detailnya
    print("Daftar Buku:")
    for buku in daftar_buku:
        print(f"Judul: {buku.judul}")
        print(f"Pengarang: {buku.pengarang}")
        print(f"Tahun Terbit: {buku.tahun_terbit}")
        print(f"Stok: {buku.jumlah_stok}")
        print("-" * 20)

# fungsi untuk mencari buku
def cari_buku(judul):
    # Cari buku dalam daftar
    for buku in daftar_buku:
        if buku.judul.lower() == judul.lower():
            # Jika ditemukan, cetak detail buku
            print(f"Judul: {buku.judul}")
            print(f"Pengarang: {buku.pengarang}")
            print(f"Tahun Terbit: {buku.tahun_terbit}")
            print(f"Jumlah Stok: {buku.jumlah_stok}")
            return
    # Jika tidak ditemukan, keluarkan pesan error buku tidak ditemukan
    raise ValueError(f"Buku {judul} tidak ditemukan.")
# Untuk menampilkan menu dan menerima input pengguna
while True:
    print(" ")
    print("Menu:")
    print("1. Tambah Buku Baru")
    print("2. Pinjam Buku")
    print("3. Kembalikan Buku")
    print("4. Cari Buku")
    print("5. Tampilkan Semua Buku")
    print("6. Selesai")
    print(" ")

    pilihan = input("Pilih menu yang Anda inginkan (1/2/3/4/5/6): ")

    try:
    # fungsi untuk menambahkan buku baru ke dalam daftar
        if pilihan == "1":
            judul = input("Judul Buku: ")
            pengarang = input("Nama Pengarang: ")
            tahun_terbit = int(input("Tahun Terbit: "))
            try:
                jumlah_stok = int(input("Jumlah Stok: "))
            except ValueError:
                # Jika input stok bukan angka, maka cetak pesan error jumlah stok harus berupa angka
                print("Error! Jumlah stok harus berupa angka.")
                continue
            buku_baru = Buku(judul, pengarang, tahun_terbit, jumlah_stok)
            daftar_buku.append(buku_baru)
            data_buku[judul] = buku_baru
            print(f"Buku {judul} berhasil ditambahkan.")
    # fungsi untuk meminjam buku berdasarkan judul
        elif pilihan == "2":
            judul = input("Judul buku yang ingin dipinjam: ")
            if judul in data_buku:
                data_buku[judul].pinjam()
            else:
              # Jika buku tidak ditemukan, maka menampilkan pesan error buku tidak ditemukan
              raise ValueError(f"Buku {judul} tidak ditemukan.")
    # fungsi untuk mengembalikan buku berdasarkan judul
        elif pilihan == "3":
            judul = input("Judul buku yang ingin dikembalikan: ")
            if judul in data_buku:
                data_buku[judul].kembalikan()
            else:
              # Jika buku tidak ditemukan, maka menampilkan pesan error buku tidak ditemukan
              raise ValueError(f"Buku {judul} tidak ditemukan.")
    # fungsi untuk mencari buku berdasarkan judul
        elif pilihan == "4":
            judul = input("Masukkan judul buku yang ingin dicari: ")
            cari_buku(judul)
    # fungsi untuk menampilkan semua buku yang ada dalam daftar
        elif pilihan == "5":
          tampilkan_semua_buku()
    # fungsi untuk keluar dari program
        elif pilihan == "6":
            print("Terima kasih! Sampai jumpa!")
            break
        else:
            # jika pilihan tidak valid, maka menampilkan pesan pilihan tidak valid. Silakan coba lagi
            print("Pilihan tidak valid. Silakan coba lagi.")
    except ValueError as e:
        # Untuk menangani error yang disebabkan ValueError
        print(f"Mohon maaf. {e}")
    except Exception as e:
        # Untuk menangani error lain
        print(f"Terjadi kesalahan: {e}")
```
